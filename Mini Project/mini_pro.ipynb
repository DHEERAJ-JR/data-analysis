### plotting total unemployment - From 1991 up to 2014

```python
plt.figure(figsize=(15,8))
plt.plot(iv1['Year'].values,iv1['Value'].values,'-gD')
#plt.plot(nv1['Year'].values,nv1['Value'].values,'-bD')
plt.axis([1989, 2015,0,6])

for x,y in zip(iv1['Year'].values,iv1['Value'].values):
    label = "{:.2f}".format(y)
    plt.annotate(label,(x,y),textcoords="offset points",xytext=(0,10),ha='center')

plt.grid(True)
plt.ylabel('Unemployment, total%')
plt.xlabel('years')
plt.title('Total Unemployment - From 1991 up to 2014')
```
